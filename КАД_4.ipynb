```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 600
n_interval = 15
```

```python
np.random.seed(1)
data = np.zeros(n)
data = np.random.normal(loc = 0, scale = 1, size = int(n))
```

```python
data_interval = np.zeros([15,40])
for i in range(n_interval):
  data_interval[i] = data[i*40:(i+1)*40]
```

```python
fig = plt.figure(figsize=(16, 10), dpi=60, facecolor="w", edgecolor="k")

plt.plot(data, c='b', marker='o', alpha = 0.8)
plt.grid(True)

None
```

```python
mean_data = np.mean(data_interval, axis =1)
mean_data
```

```python
var_data = np.var(data_interval, axis =1)
var_data
```

```python
fig = plt.figure(figsize=(16, 10), dpi=60, facecolor="w", edgecolor="k")

plt.plot(np.arange(len(mean_data)), mean_data, c='b', marker='o', alpha = 0.8)
plt.hlines(np.median(mean_data), 0, 15, color ='r')
plt.xlabel(' ', fontsize=18)
plt.ylabel(' ', fontsize=18)
plt.xticks(fontsize=14, fontweight=1000)
plt.yticks(fontsize=14, fontweight=1000)
plt.grid(True)

None
```

```python
series_method_result = mean_data > np.median(mean_data)
series_method_result
```

```python
mean_A_data = np.zeros(len(data_interval))
var_A_data = np.zeros(len(data_interval))

for sample in range(len(data_interval)):
  for element in range(len(data_interval)):
    if mean_data[sample] > mean_data[element]:
      mean_A_data[sample] += 1

for sample in range(len(data_interval)):
  for element in range(len(data_interval)):
    if var_data[sample] > var_data[element]:
      var_A_data[sample] += 1

mean_A_data = mean_A_data.sum()

var_A_data = var_A_data.sum()
```
